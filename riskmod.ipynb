```python
import riskmod, math
import matplotlib.pyplot as plt
v, a = riskmod.run()
x = [i for i in range(-8, 11)]
plt.bar(x, [i / 100000 for i in v])
plt.xticks(x)
print()
plt.show()
```

```python
expectx = 0
for i in range(-8, 11):
    expectx += i * v[i + 8] / 100000
print(expectx)
```

```python
print(a / 100000)
```

```python
stddev = 0
for i in range(0, 19):
    stddev += (expectx - v[i]) * (expectx - v[i])
print(math.sqrt(stddev))
```
