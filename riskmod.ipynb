```python
import riskmod, math
import matplotlib.pyplot as plt
trials = 10000000
```

```python
v, a = riskmod.run(trials)
print(v)
print(a)
```

```python
x = [i for i in range(-8, 11)]
plt.bar(x, [i / trials for i in v])
plt.xticks(x)
plt.show()
```

```python
expectx = 0
for i in range(-8, 11):
    expectx += i * v[i + 8] / trials
print(expectx)
```

```python
print(a / trials)
```

```python
stddev = 0
for i in range(-8, 11):
    stddev += v[i + 8] * (expectx - i) * (expectx - i)
print(math.sqrt(stddev / trials))
```
